```python
df=pd.DataFrame({'name':['Tom','Andy','lucas','pranav','uma','rahul','ashwin','kashish'],
                 'dob':['08/23/2000','03/06/2002','09/07/2003','12/20/2001','08/23/2005','11/04/2002','07/12/2007','02/18/2004']})
df['dob']=pd.to_datetime(df['dob'])
df
```

```python
today=pd.to_datetime('today')
df['age']=today.year-df['dob'].dt.year
df
```

```python
df['year']=df['dob'].dt.year
df['month']=df['dob'].dt.month
df['date']=df['dob'].dt.date
df
```

```python
df=df.set_index(['dob'])
df
```
